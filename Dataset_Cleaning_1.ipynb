Step 1: Dataset CLEAN

```python
# Load dataset
import pandas as pd

df = pd.read_csv("online_retail_dataset.csv")
```

```python
df.info()
```

```python
#Basic cleaning
df = df.dropna(subset=['CustomerID'])

df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]

df['InvoiceDate'] = pd.to_datetime(
    df['InvoiceDate'],
    errors='coerce',
    dayfirst=True
)

df = df.dropna(subset=['InvoiceDate'])

df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

STEP 2: Customer-based Data Reduction

```python
# how many customers?
print("Total Customers:", df['CustomerID'].nunique())

# select customers
customers = df['CustomerID'].unique()

selected_customers = pd.Series(customers).sample(
    n=3000,
    random_state=42
)

# keep full history
df_reduced = df[df['CustomerID'].isin(selected_customers)]

print("Reduced Rows:", df_reduced.shape[0])
print("Reduced Customers:", df_reduced['CustomerID'].nunique())
```

Save Reduced Dataset 

```python
df_reduced.to_csv("retail_reduced_customers.csv", index=False)
```
